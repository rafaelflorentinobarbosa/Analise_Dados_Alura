# 1. Análise do faturamento


## a) Faturamento total das lojas com o frete

```python
faturamento_loja1 = loja["Preço"].sum()
faturamento_loja2 = loja2["Preço"].sum()
faturamento_loja3 = loja3["Preço"].sum()
faturamento_loja4 = loja4["Preço"].sum()
```

```python
print(f"Faturamento da Loja 1: R$ {faturamento_loja1:.2f}")
print(f"Faturamento da Loja 2: R$ {faturamento_loja2:.2f}")
print(f"Faturamento da Loja 3: R$ {faturamento_loja3:.2f}")
print(f"Faturamento da Loja 4: R$ {faturamento_loja4:.2f}")
```

```python
import matplotlib.ticker as mticker

dados_faturamento = {
    "Loja 1": faturamento_loja1,
    "Loja 2": faturamento_loja2,
    "Loja 3": faturamento_loja3,
    "Loja 4": faturamento_loja4,
}

# Convertendo o dicionário em uma Series do Pandas para facilitar a plotagem.
faturamentos_series = pd.Series(dados_faturamento)

# --- Criação do gráfico de barras ---
plt.figure(
    figsize=(12, 7)
)  # Aumenta um pouco o tamanho da figura para melhor visualização
ax = faturamentos_series.plot(
    kind="bar",  # Especifica que o tipo de gráfico será de barras
    color=[
        "skyblue",
        "lightcoral",
        "lightgreen",
        "gold",
    ],  # Define cores para as barras
)

# --- Personalização do gráfico ---
plt.title(
    "Faturamento Total por Loja", fontsize=18, fontweight="bold"
)  # Título do gráfico com fonte maior e negrito
plt.xlabel("Loja", fontsize=14)  # Rótulo do eixo X
plt.ylabel(
    "Faturamento", fontsize=14
)  # Rótulo do eixo Y (o R$ será adicionado na formatação)
plt.xticks(
    rotation=0, fontsize=12
)  # Garante que os rótulos do eixo X fiquem horizontais e com fonte maior
plt.yticks(fontsize=12)  # Define o tamanho da fonte para os rótulos do eixo Y
plt.grid(
    axis="y", linestyle="--", alpha=0.7
)  # Adiciona uma grade horizontal para facilitar a leitura

# --- Formatação dos rótulos do eixo Y como moeda (R$) ---
# Define uma função de formatação para o eixo Y.
# Ela adiciona "R$" e formata o número com separador de milhares e duas casas decimais.
formatter = mticker.FormatStrFormatter("R$ %.2f")
ax.yaxis.set_major_formatter(formatter)

# Adiciona os valores de faturamento em cima de cada barra
for i, v in enumerate(faturamentos_series.values):
    ax.text(
        i,
        v + (faturamentos_series.max() * 0.02),
        f"R$ {v:,.2f}".replace(",", "X").replace(".", ",").replace("X", "."),
        color="black",
        ha="center",
        va="bottom",
        fontsize=10,
        fontweight="bold",
    )

plt.tight_layout()  # Ajusta automaticamente os parâmetros da subtrama para que se encaixem na área da figura.

plt.show()
```

## Faturamento Mensal

```python
# Leitura dos dados
loja1 = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

# Função para preparar os dados de cada loja
def preparar_dados(loja):
    loja['Data da Compra'] = pd.to_datetime(loja['Data da Compra'], dayfirst=True)
    loja['Rendimento'] = loja['Preço']  # Se rendimento for apenas preço da venda
    loja['Ano-Mes'] = loja['Data da Compra'].dt.to_period('M').astype(str)
    media_rendimento = loja.groupby('Ano-Mes')['Rendimento'].mean()
    return media_rendimento

# Preparar dados de todas as lojas
rendimento_loja1 = preparar_dados(loja1)
rendimento_loja2 = preparar_dados(loja2)
rendimento_loja3 = preparar_dados(loja3)
rendimento_loja4 = preparar_dados(loja4)

# Criar o gráfico
plt.figure(figsize=(12, 6))

plt.plot(rendimento_loja1.index, rendimento_loja1.values, marker='o', label='Loja 1')
plt.plot(rendimento_loja2.index, rendimento_loja2.values, marker='s', label='Loja 2')
plt.plot(rendimento_loja3.index, rendimento_loja3.values, marker='^', label='Loja 3')
plt.plot(rendimento_loja4.index, rendimento_loja4.values, marker='d', label='Loja 4')

plt.xlabel('Ano-Mês')
plt.ylabel('Rendimento Médio (R$)')
plt.title('Rendimento Médio Mensal por Loja')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)

plt.tight_layout()
plt.show()
```

## Faturamento Anual

```python
# Leitura dos dados
loja1 = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

# Função para preparar os dados anuais de cada loja
def preparar_dados_anuais(loja):
    loja['Data da Compra'] = pd.to_datetime(loja['Data da Compra'], dayfirst=True)
    loja['Rendimento'] = loja['Preço']  # Ou preço * quantidade, se desejar
    loja['Ano'] = loja['Data da Compra'].dt.year
    media_anual = loja.groupby('Ano')['Rendimento'].mean()
    return media_anual

# Preparar dados de todas as lojas
rendimento_loja1 = preparar_dados_anuais(loja1)
rendimento_loja2 = preparar_dados_anuais(loja2)
rendimento_loja3 = preparar_dados_anuais(loja3)
rendimento_loja4 = preparar_dados_anuais(loja4)

# Criar o gráfico
plt.figure(figsize=(10, 6))

plt.plot(rendimento_loja1.index, rendimento_loja1.values, marker='o', label='Loja 1')
plt.plot(rendimento_loja2.index, rendimento_loja2.values, marker='s', label='Loja 2')
plt.plot(rendimento_loja3.index, rendimento_loja3.values, marker='^', label='Loja 3')
plt.plot(rendimento_loja4.index, rendimento_loja4.values, marker='d', label='Loja 4')

plt.xlabel('Ano')
plt.ylabel('Rendimento Médio (R$)')
plt.title('Rendimento Médio Anual por Loja')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

# 2. Vendas por Categoria


```python
# Função para calcular a quantidade de produtos vendidos por categoria e as mais populares
def analisar_categorias(df, nome_loja):
    print(f"Análise de Categorias para {nome_loja}:")

    # Agrupa por 'Categoria do Produto' e conta a ocorrência de cada uma
    contagem_categorias = df["Categoria do Produto"].value_counts()
    print(contagem_categorias)

    # Encontra a categoria mais popular
    if not contagem_categorias.empty:
        categoria_mais_popular = contagem_categorias.index[0]
        quantidade_mais_popular = contagem_categorias.iloc[0]
        print(
            f"  Categoria mais popular: '{categoria_mais_popular}' com {quantidade_mais_popular} vendas."
        )

        # Geração do gráfico de barras para as categorias mais populares
        plt.figure(figsize=(10, 6))
        contagem_categorias.head(10).plot(kind="bar", color="skyblue")
        plt.title(f"Top 10 Categorias de Produtos em {nome_loja}")
        plt.xlabel("Categoria do Produto")
        plt.ylabel("Número de Vendas")
        plt.xticks(rotation=45, ha="right")
        plt.tight_layout()
        plt.show()
        plt.close()

    else:
        print("  Não há dados de categoria para analisar.")
    print("-" * 30)
```

```python
# Chama a função para cada loja
analisar_categorias(loja, "Loja 1")
analisar_categorias(loja2, "Loja 2")
analisar_categorias(loja3, "Loja 3")
analisar_categorias(loja4, "Loja 4")
```

# 3. Média de Avaliação das Lojas

```python
avaliacao_loja1 = loja["Avaliação da compra"].mean()
avaliacao_loja2 = loja2["Avaliação da compra"].mean()
avaliacao_loja3 = loja3["Avaliação da compra"].mean()
avaliacao_loja4 = loja4["Avaliação da compra"].mean()
```

```python
print(f"Média de avaliação da Loja 1: {avaliacao_loja1:.2f}")
print(f"Média de avaliação da Loja 2: {avaliacao_loja2:.2f}")
print(f"Média de avaliação da Loja 3: {avaliacao_loja3:.2f}")
print(f"Média de avaliação da Loja 4: {avaliacao_loja4:.2f}")
```

```python
# Cálculo das médias de avaliação
medias = {
    'Loja 1': loja['Avaliação da compra'].mean(),
    'Loja 2': loja2['Avaliação da compra'].mean(),
    'Loja 3': loja3['Avaliação da compra'].mean(),
    'Loja 4': loja4['Avaliação da compra'].mean()
}

# Ordenar do maior para o menor
medias_ordenadas = dict(sorted(medias.items(), key=lambda item: item[1], reverse=True))

# Dados ordenados
lojas = list(medias_ordenadas.keys())
valores = list(medias_ordenadas.values())

# Criar gráfico de barras horizontal
plt.figure(figsize=(8, 5))
plt.barh(lojas, valores, color='skyblue')
plt.xlabel('Média de Avaliações')
plt.title('Média de Avaliações por Loja (Ordenado)')
plt.xlim(0, 5)  # Como as avaliações vão de 1 a 5

# Exibir os valores no gráfico
for index, value in enumerate(valores):
    plt.text(value + 0.05, index, round(value, 2))

plt.tight_layout()
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
# Função para plotar os Top 5 mais e menos vendidos lado a lado
def plot_top5_mais_menos(loja, nome_loja):
    contagem = loja['Produto'].value_counts()
    top5_mais = contagem.nlargest(5)
    top5_menos = contagem.nsmallest(5)

    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    # Mais vendidos
    axes[0].barh(top5_mais.index[::-1], top5_mais.values[::-1], color='lightgreen')
    axes[0].set_title(f'Top 5 Mais Vendidos - {nome_loja}')
    axes[0].set_xlabel('Quantidade')

    for index, value in enumerate(top5_mais.values[::-1]):
        axes[0].text(value + 0.5, index, value)

    # Menos vendidos
    axes[1].barh(top5_menos.index[::-1], top5_menos.values[::-1], color='salmon')
    axes[1].set_title(f'Top 5 Menos Vendidos - {nome_loja}')
    axes[1].set_xlabel('Quantidade')

    for index, value in enumerate(top5_menos.values[::-1]):
        axes[1].text(value + 0.5, index, value)

    plt.tight_layout()
    plt.show()

# Plotar para cada loja
plot_top5_mais_menos(loja, 'Loja 1')
plot_top5_mais_menos(loja2, 'Loja 2')
plot_top5_mais_menos(loja3, 'Loja 3')
plot_top5_mais_menos(loja4, 'Loja 4')
```
